```python
#we normalise our dataset to a common scale using the Min Max scaler
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
datascaled=scaler.fit_transform(df)
normalised_df=pd.DataFrame(datascaled, columns= df.columns)
```

```python
x=normalised_df.drop(columns=['Appliances'])
y=normalised_df['Appliances']
```

```python
from sklearn.linear_model import LinearRegression
mlm=LinearRegression()
from sklearn.model_selection import train_test_split

x_train,x_test, y_train,y_test=train_test_split(x,y,test_size=0.3, random_state=42)
#fit the model to the training dataset
mlm.fit(x_train,y_train)
#obtain predictions
predicted_values = mlm.predict(x_test)
#to check for overfittness and underfitness
y_pred_train=mlm.predict(x_train)
from sklearn.metrics import r2_score
y_pred_test=predicted_values
train_score=r2_score(y_train,y_pred_train)
test_score=r2_score(y_test,y_pred_test)
print(train_score)
print(test_score)
```

```python
#comparing the effects of regularisation
def get_weights_df(model,feat,col_name):
    #this function returns the weight of every feature
    weights= pd.Series(model.coef_,feat.columns).sort_values()
    weights_df=pd.DataFrame(weights).reset_index()
    weights_df.columns=['Features',col_name]
    weights_df[col_name].round(3)
    return weights_df
linear_model_weights=get_weights_df(Lasso_reg,x_train,'linear_model_weights')
linear_model_weights
```

```python
ridge_weights_df=get_weights_df(Ridge_reg, x_train, 'Ridge_weight')
ridge_weights_df
lasso_weights_df=get_weights_df(Lasso_reg,x_train, 'Lasso_weight')
lasso_weights_df
final_weights=pd.merge(linear_model_weights,lasso_weights_df, on='Features')
final_weights=pd.merge(final_weights,ridge_weights_df, on='Features')
final_weights
```
